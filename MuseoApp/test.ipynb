```python
import requests

# Definir la URL base de la API
base_url = 'http://localhost:5000'  # Ajusta el puerto según donde se esté ejecutando tu app Flask

# Función para hacer una petición GET y mostrar la respuesta
def get_data_and_display(endpoint):
    url = f"{base_url}/{endpoint}"
    response = requests.get(url)
    if response.status_code == 200:
        print(f"Data from {url}:")
        print(response.json())
    else:
        print(f"Failed to fetch data from {url}. Status code: {response.status_code}")

# Ejemplo de llamadas a los endpoints
endpoints = [
    'artefacts/exhibitions',
    'artefacts/1/exhibitions',
    'artefacts/exhibitions/between_dates?start_date=2023-01-01&end_date=2023-12-31',
    'artefacts/1/exhibitions/between_dates?start_date=2023-01-01&end_date=2023-12-31',
    'artefacts/ordered_by_family_and_name',
    'expositions_and_artefacts_ordered_by_start_date',
    'expositions_and_total_tours',
    'expositions_and_total_tours_between_dates?start_date=2023-01-01&end_date=2023-12-31',
    'expositions_and_visiting_companies'
]

# Realizar las peticiones a cada endpoint y mostrar los resultados
for endpoint in endpoints:
    get_data_and_display(endpoint)
```
